**Decision Tree ve Regression: Çalışma Prensibi**

 - Karar ağaçları, hem classfication hem de regression problemlerinde etkili bir şekilde kullanılan güçlü ML modellerindendir.   - Regression problemleri için kullanıldığında, decision tree nin temel çalışma prensibi, veri seti içindeki sürekli değişkenlerin değerlerini tahmin etmek üzerine kuruludur. 
 - Bu süreç, veri setinin homojen alt kümelerine bölünmesiyle gerçekleşir, böylece her alt kümenin varyansı en aza indirgenir. 

**Variance ve Mean Temelli Yaklaşım**

**Farklılık**
 - Decision Tree, regression problemlerinde, classification problemlerinde yaygın olarak kullanılan entropi veya gini gibi saflık ölçütlerine başvurmaz. 
 - Bunun yerine, ağaç yapıları, veri seti içindeki sürekli değişkenlerin tahminlerini iyileştirmek için varyans minimizasyonu gibi istatistiksel yöntemleri kullanır.

**Varyans Kullanımı**
 - Decision Tree Regressor, veri setini bölerken, her bir bölme işlemi sonrasında elde edilen alt kümelerin varyansını minimize edecek şekilde tasarlanmıştır. 
 - Varyans, bir veri setindeki değerlerin, veri setinin ortalamasından ne kadar farklılık gösterdiğinin bir ölçüsüdür. 
 - Bir karar ağacı, varyansı minimize ederek, benzer değerlere sahip veri noktalarını bir araya getirir ve böylece daha tutarlı tahminler yapılmasını sağlar.

**Tahmin Yöntemi**

 - Decision Tree Regressor'de, bir dalın veya düğümün kapsadığı veri noktalarının ortalaması, o dal veya düğüm için tahmini değer olarak kullanılır.
 - Bu yöntem, belirli bir bölgedeki tüm veri noktalarının değerlerinin ortalamasını alarak, o bölge için genelleştirilmiş bir tahmin sunar.

**İleri Düzey Bilgiler ve İyileştirmeler**

- **Pruning**: Karar ağaçlarının overfitting eğilimini azaltmak için, pruning teknikleri kullanılabilir. Pruning, ağacın gereksiz dallarını kırparak modelin genelleştirme yeteneğini artırır.
- **Cross Validation**: Modelin performansını değerlendirmek ve overfitting riskini minimize etmek için cross validation teknikleri kullanılabilir. Bu, modelin farklı veri alt kümeleri üzerindeki performansını test ederek, daha sağlam ve güvenilir tahminler yapmasına yardımcı olur.
- **Feature Importance**: Decision Tree, hangi feature ların tahminler üzerinde en fazla etkiye sahip olduğunu belirlemede kullanışlıdır. Bu, modelin hangi feature ların hedef değişken üzerinde en güçlü etkiye sahip olduğunu anlamak için önemli bir araçtır.

Decision Tree Regressor, bu prensipler ve teknikler doğrultusunda, veri setlerindeki karmaşık ilişkileri ve desenleri modelleyebilir ve sürekli değişkenlerin değerlerini etkili bir şekilde tahmin edebilir. 

Bu modeller, özellikle farklı özellikler arasındaki ilişkilerin doğrusal olmadığı durumlarda, güçlü ve esnek tahmin araçları olarak öne çıkar.

# Modelling

## Train | Test Split

```python
X = df.drop("Selling_Price", axis=1)
y = df.Selling_Price
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.2,
                                                    random_state=101)

print("Train features shape : ", X_train.shape)
print("Train target shape   : ", y_train.shape)
print("Test features shape  : ", X_test.shape)
print("Test target shape    : ", y_test.shape)
```

## Modelling with Pipeline & ColumnTransformer

```python
cat = X_train.select_dtypes("object").columns
cat
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OrdinalEncoder

ord_enc = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
column_trans = make_column_transformer((ord_enc, cat), remainder='passthrough')
```

```python
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor

operations = [("OrdinalEncoder", column_trans),
              ("DT_model", DecisionTreeRegressor(random_state=101))]

pipe_model = Pipeline(steps=operations)

pipe_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
def train_val(model, X_train, y_train, X_test, y_test):

    y_pred = model.predict(X_test)
    y_train_pred = model.predict(X_train)

    scores = {
        "train": {
            "R2": r2_score(y_train, y_train_pred),
            "mae": mean_absolute_error(y_train, y_train_pred),
            "mse": mean_squared_error(y_train, y_train_pred),
            "rmse": np.sqrt(mean_squared_error(y_train, y_train_pred))
        },
        "test": {
            "R2": r2_score(y_test, y_pred),
            "mae": mean_absolute_error(y_test, y_pred),
            "mse": mean_squared_error(y_test, y_pred),
            "rmse": np.sqrt(mean_squared_error(y_test, y_pred))
        }
    }

    return pd.DataFrame(scores)
```

```python
train_val(pipe_model, X_train, y_train, X_test, y_test)
```

```python
train_val(pipe_model, X_train, y_train, X_test, y_test)

# overfiting olduğuna dair insightımızı aldık. CV'da kesin kararımızı vereceğiz.
```

```python
from sklearn.model_selection import cross_validate, cross_val_score

operations = [("OrdinalEncoder", column_trans),
              ("DT_model", DecisionTreeRegressor(random_state=101))]
model = Pipeline(steps=operations)

scores = cross_validate(model,
                        X_train,
                        y_train,
                        scoring=[
                            'r2', 'neg_mean_absolute_error',
                            'neg_mean_squared_error',
                            'neg_root_mean_squared_error'
                        ],
                        cv=10,
                        return_train_score=True)

df_scores = pd.DataFrame(scores)
df_scores.mean()[2:]

# overfiting olduğunu teyit ettik
```

## Overfitting Risk and Features Importance

**Decision Tree Modellerde Overfitting Analizi**

**Eğitim Verileri Üzerindeki Model Performansının İncelemesi**

- **Gözlem**: Eğer bir decision tree modeli eğitim veri setinde neredeyse veya tamamen hatasız performans gösteriyorsa, bu durum modelin eğitim verilerine karşı overfitting eğilimi gösterdiğinin bir göstergesi olabilir.
- **Analiz**: Eğitim setindeki yüksek performansın analizi, modelin ayrıntıları yakalarken esas örüntüleri kaçırma riski taşıdığını ortaya koyar. 

**Cross Validation Sonuçlarının Değerlendirilmesi**

- **Gözlem**: Modelin cross validation skorlarının, eğitim setinde elde edilen skorlardan önemli ölçüde düşük olması, modelin genelleştirme yeteneğinin zayıf olduğuna işaret eder.
- **Analiz**: Cross validation, modelin farklı veri alt kümelerindeki performansını değerlendirmek için güvenilir bir yöntemdir. Eğer cross validation skorları eğitim skorlarından düşükse, bu durum overfitting belirtisi olarak değerlendirilir.

**Overfitting Durumunun Giderilmesi için Öneriler**

- **Hiperparametre Ayarlaması**: Modelin karmaşıklığını azaltmak ve genelleştirme yeteneğini artırmak için hiperparametrelerin ayarlanması gerekebilir. Örneğin, karar ağacının derinliğinin sınırlandırılması, budama yapılması gibi teknikler kullanılabilir.
- **Veri Setinin Genişletilmesi**: Modelin genelleştirme kapasitesini artırmak için daha fazla veri ile eğitilmesi sağlanabilir. Bu, modelin daha çeşitli örnekleri görmesini ve daha güçlü örüntüleri öğrenmesini sağlar.
- **Feature Engineering**: Modelin genelleştirme performansını etkileyebilecek gereksiz veya yanıltıcı özelliklerin çıkarılması, önemli feature ların seçilmesi ve dönüştürülmesi de faydalı olabilir.

```python
operations = [("OrdinalEncoder", column_trans),
              ("DT_model", DecisionTreeRegressor(random_state=101))]
pipe_model = Pipeline(steps=operations)
pipe_model.fit(X_train, y_train)

scores = cross_validate(pipe_model,
                        X_train,
                        y_train,
                        scoring=[
                            'r2', 'neg_mean_absolute_error',
                            'neg_mean_squared_error',
                            'neg_root_mean_squared_error'
                        ],
                        cv=10,
                        return_train_score=True)

df_scores = pd.DataFrame(scores)
df_scores.mean()[2:]
```

```python
pipe_model["DT_model"].feature_importances_
```

```python
features = pipe_model["OrdinalEncoder"].get_feature_names_out()
features
```

```python
new_features = [
    i.replace("ordinalencoder__", "").replace("remainder__", "")
    for i in features
]
new_features

# ordinalencoder__ ve remainder__ ön eklerinden kurtuluyoruz.
```

```python
df_fi = pd.DataFrame(
    data=pipe_model["DT_model"].feature_importances_,
    index=new_features,  #index=X.columns
    columns=["Feature Importance"])
df_fi = df_fi.sort_values("Feature Importance", ascending=False)

df_fi

# featue importance oranlarını gösteren df'i df_fi olarak tanımladık.
# df içerisine pipe_model["DT_model"].feature_importances_ ile feature importance oranlarını, make_column_transform
# sonrası sıralaması değişen feature isimleri (new_features) index'e tanımlıyoruz.
# Sonra da feature importance değerlerini  büyükten küçüğe doğru sıralıyoruz.
```

```python
ax = sns.barplot(x=df_fi.index, y='Feature Importance', data=df_fi)
ax.bar_label(ax.containers[0], fmt="%.3f")
plt.xticks(rotation=90)
plt.show()

# feature importance değerlerini barplot ile görselleştiriyoruz.
# Feature selection yaparken burdaki sıralamaya göre yapıyoruz.
```

**Feature Importance Sınırlamaları ve İncelemesi**

- **Sınırlama ve Önem Derecelendirmesi**: Karar ağaçlarında feature importance, genellikle o feature ın ağaçta kullanılma sıklığı ve yaprak düğümlerinde sağladığı ortalama bilgi kazancına dayanır. Ancak bu, her feature ın modelin tahminlerine katkısının mutlak değerini doğru bir şekilde yansıtmayabilir.
- **Analiz Yöntemleri**: Modelden bir feature ı çıkarmak ve modelin performansını yeniden değerlendirmek, o feature ın model üzerindeki etkisini daha net anlamak için kullanılabilir. Eğer bir feature ın çıkarılması model performansında büyük bir düşüşe neden oluyorsa, bu feature ın model için önemli olduğu sonucuna varılabilir.

**Uygulama Örneği ve Değerlendirme**

## Visualizing Trees

**Decision Tree Reggression'da Bölünme Süreci**


- **İstatistiksel Yaklaşım**: Regresyon ağaçlarında bölme işlemi, sürekli değişkenlerdeki optimal bölünme noktalarını belirlemek için titiz bir istatistiksel analiz gerektirir. Bu analiz, her potansiyel bölme noktasında varyans azaltma potansiyelini değerlendirir.

- **Bölme Kriteri**: Bölünme sürecinde kullanılan kriter, genellikle Minimum Varyans Redüksiyonu (Minimum Variance Reduction - MVR) veya Ortalama Karesel Hata Azaltma (Mean Squared Error Reduction - MSER) gibi bir metriktir. Bu metrikler, veri setinin homojen alt gruplara ayrılmasını sağlayarak, her bir grup içindeki değerlerin ortalama etrafında daha sıkı bir dağılım göstermesine yardımcı olur.

- **Optimizasyon**: Bölme işlemi, varyansı en aza indirecek şekilde veri setinin sürekli değerlerini gruplara ayırır. Bu gruplar içindeki veri noktalarının ortalaması, o gruba düşen gözlemlerin tahmin değeri olarak kullanılır. Bu, modelin tahminlerini veri setinin gerçek dağılımına daha uygun hale getirir.

- **Modelin Davranışı**: Regresyon ağacı, sürekli değişkenler üzerinde kesintisiz bir bölme süreci uygular ve bu süreç, her bir alt grubun ortalamasını kullanarak tahminler yapar. Bu metodoloji, modelin eğitim veri setindeki değişkenlikleri ve desenleri etkili bir şekilde yakalamasını ve yeni veriler üzerindeki tahminlerini iyileştirmesini sağlar.

**Etkinlik ve Karmaşıklık**

- **Ağacın Büyüklüğü**: Regresyon ağaçları genellikle sınıflandırma ağaçlarına göre daha fazla düğüm ve dal oluşturabilir, çünkü sürekli değişkenlerin bölünme noktaları daha fazla ve daha detaylı olabilir. Bu, modelin karmaşıklığını artırabilir ve overfitting riskini beraberinde getirebilir.

- **Pruning Stratejileri**: Bu karmaşıklığı yönetmek için pruning stratejileri devreye girer. Pruning işlemi, modelin genelleştirme yeteneğini korurken aşırı karmaşıklığın önüne geçer. Modelin overfitting meyilli olmasını önleyerek, genelleştirme hatasını azaltmaya yardımcı olur.

## Grid Search

**criterion**

1. **squared_error**:
   - Bu, bölünme kalitesini değerlendirmek için kullanılan ortalama karesel hata (MSE) kriteridir.
   - Her düğümde, ortalama karesel hatayı minimize etmeye çalışarak en iyi bölünmeyi bulur.
   - Bu, genellikle varsayılan kriterdir ve ağacın hedef değişken üzerindeki varyansı azaltmasını sağlar.

2. **friedman_mse**:
   - Bu kriter, ünlü istatistikçi Jerome Friedman tarafından geliştirilen ve geliştirilmiş bir karesel hata ölçüsüdür.
   - 'friedman_mse' genellikle gradyan artırma ağaçlarıyla (gradient boosting trees) kullanılır ve ağacın bölünmelerini iyileştirmek için ek bir düzeltme terimi içerir.
   - Bu kriter, özellikle gradyan artırma modellerindeki performansı artırmak için kullanılır.

3. **absolute_error**:
   - Bu kriter, bölünme kalitesini belirlemek için ortalama mutlak hatayı (MAE) kullanır.
   - Aykırı değerlerin etkisini azaltmak için tercih edilen bir kriterdir çünkü karesel değil, mutlak farkları hesaba katar.
   - Hesaplama açısından daha maliyetli olabilir, çünkü mutlak değer fonksiyonunun türevi her noktada tanımlı değildir.

4. **poisson**:
   - Bu kriter, regresyon hedeflerinin Poisson dağılımı gösterdiği durumlar için uygundur. Örneğin, sayım verisi veya olay oranları gibi.
   - Poisson kaybı, bir hedef değişkenin gerçek sayımlar üzerindeki tahminlerinin kalitesini değerlendirmek için kullanılır.
   - Poisson regresyonu, özellikle hedef değişken bir sayma sayısı olduğunda (örneğin, bir zaman aralığında gerçekleşen olayların sayısı gibi) ve hata terimlerinin varyansı ortalama değerlerle orantılı olduğunda kullanılır.
